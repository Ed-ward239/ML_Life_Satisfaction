```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mpl
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

mpl.rc('axes', labelsize=17)
mpl.rc('xtick', labelsize=17)
mpl.rc('ytick', labelsize=17)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CUNY-CISC3440-Fall2022/datasets/main/life_satisfaction.csv')
```

```python
df['gdp_per_capita']
```

```python
model = LinearRegression()

test = df.iloc[[0]]
train = df.drop(0)


X = train[['gdp_per_capita']]
y = train['life_satisfaction']

model.fit(X, y)

y_pred = model.predict(test[['gdp_per_capita']])
mean_squared_error(test['life_satisfaction'], y_pred, squared=False)
```

```python
# P1 (1): Mean RSME
model = LinearRegression()

num = len(df)
lst = []

for x in range(len(df)):
  test = df.iloc[[x]]
  train = df.drop(x)
  X = train[['gdp_per_capita']]
  y = train['life_satisfaction']
  model.fit(X, y)
  y_pred = model.predict(test[['gdp_per_capita']])
  error = mean_squared_error(test['life_satisfaction'], y_pred, squared=False)
  error
  lst.append(error)

np.mean(lst)
```

```python
# P1 (2): Mean K-Nearest-Neighbors Regressor
import sklearn.neighbors
model = sklearn.neighbors.KNeighborsRegressor(n_neighbors=3)

num = len(df)
lst = []

for x in range(len(df)):
  test = df.iloc[[x]]
  train = df.drop(x)
  X = train[['gdp_per_capita']]
  y = train['life_satisfaction']
  model.fit(X, y)
  y_pred = model.predict(test[['gdp_per_capita']])
  error = mean_squared_error(test['life_satisfaction'], y_pred, squared=False)
  error
  lst.append(error)

np.mean(lst)
```
